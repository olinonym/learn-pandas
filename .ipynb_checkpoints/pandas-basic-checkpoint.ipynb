```python
# import library
import pandas as pd
```

```python
# read csv
df = pd.read_csv("train.csv") # อ่านไฟล์ CSV เป็น DataFrame
df
```

```python
# dataframe
df.shape # จำนวนแถวและคอลัมน์
df.columns # ชื่อคอลัมน์
df.index # ช่วง
df.dtypes # ชนิดข้อมูลแต่ละคอลัมน์
```

```python
# data access
df.loc[0] # เลือกแถวโดย index
df.iloc[0] # เลือกแถวโดยตำแหน่ง
df.head(5) # 5 แถวแรก
df.tail(5) # 5 แถวท้าย
```

```python
# missing NA
df.dropna() # ลบแถวที่มีค่า NA
df.dropna(axis=1) # ลบคอลัมน์ที่มีค่า NA
```

```python
# replace NA
df.fillna(0) # แทนด้วย 0
df.fillna(df.mean()) # แทนด้วยค่าเฉลี่ยของคอลัมน์
```

```python
# filter / conditional selection
df[df['Age'] > 30]
df[(df['Age'] > 30) & (df['Sex'] == 'male')]
```

```python
# group by
df.groupby('Pclass')['Age'].mean() # เฉลี่ยอายุตามคลาส
df.groupby('Pclass').count()
```

```python
# basic statistics
df['Age'].mean() # ค่าเฉลี่ย
df['Age'].count() # นับจำนวนข้อมูลไม่เป็น NA
df.describe() # สรุปสถิติแบบครบ (count, mean, std, min, 25%, 50%, 75%, max)
```
